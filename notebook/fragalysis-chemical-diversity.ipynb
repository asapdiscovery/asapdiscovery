## actually generate vector arrays

```python
def fp_to_vector_array(fp):
    integer = int(fp.ToHexString()[:-1], base=16)
    binary_str = f"{integer:04096b}"
    vector_array = np.array(list(map(int, binary_str)))
    return vector_array
```
